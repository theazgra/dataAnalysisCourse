```python
%config IPCompleter.greedy=True
import pandas as pd
import sklearn.cluster
import sklearn.metrics
import matplotlib.pyplot as plot
import numpy as np
import scipy
```

```python
bars = pd.read_csv("data/bars.csv")
bars.shape
```

```python
barsSvd = np.linalg.svd(bars,full_matrices=False)
```

```python
# leftSingularVectors and rightSingularVectors are basis
leftSingularVectors = barsSvd[0]
singularValues = np.diag(barsSvd[1])
rightSingularVectors = barsSvd[2]
```

```python
barsApproximated = leftSingularVectors.dot(singularValues.dot(rightSingularVectors))
```

```python
np.linalg.norm(bars-barsApproximated,ord='fro')
```

```python
nmfStats = []
for featureCount in range(5,50):
    model = sklearn.decomposition.NMF(n_components=featureCount, init='random')
    modelWeights = model.fit_transform(bars)
    modelValues = model.components_
    barsApproximatedNmf = np.dot(modelWeights, modelValues)
    fromNorm = np.linalg.norm(bars-barsApproximatedNmf,ord='fro')
    nmfStats.append({"Feature count":featureCount, "Norm":fromNorm})
```

```python
pd.DataFrame(nmfStats)
```

```python
model = sklearn.decomposition.NMF(n_components=15, init='random')
modelWeights = model.fit_transform(bars)
modelValues = model.components_
barsApproximatedNmf = np.dot(modelWeights, modelValues)
inverse = model.inverse_transform(modelWeights)
calculatedNorm = np.linalg.norm(bars-barsApproximatedNmf,ord='fro')
inverseNorm = np.linalg.norm(bars-inverse,ord='fro')
```
