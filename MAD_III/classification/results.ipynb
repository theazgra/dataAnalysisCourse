```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
dfTree = pd.read_csv('classification/DecisionTreeCache.csv', index_col=0)
dfNN = pd.read_csv('classification/mplc.csv', index_col=0)
dfSVC = pd.read_csv('classification/svm.csv', index_col=0)
df = pd.concat([dfTree,dfNN,dfSVC]).reset_index()
# df = df.set_index("Dataset")
df
```

```python
df2 = df.iloc[::3, :]
df2 = df2.reset_index()
df2 = df2.drop(['index'],axis=1)
df2
```

```python
plt.title('Fit Time (seconds)')
ax = sns.heatmap(df2.pivot(index='Dataset', columns='Classifier', values='FitTime'), annot=True, cmap="YlOrRd")
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

```python
plt.title('Prediction Time (seconds)')
ax = sns.heatmap(df2.pivot(index='Dataset', columns='Classifier', values='PredictionTime'), annot=True, cmap="YlOrRd")
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

```python
plt.title('Precision')
ax = sns.heatmap(df2.pivot(index='Dataset', columns='Classifier', values='Precision'), annot=True, cmap="RdYlGn")
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

```python
plt.title('F1 - Score')
ax = sns.heatmap(df2.pivot(index='Dataset', columns='Classifier', values='F1Score'), annot=True, cmap="RdYlGn")
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```
