```python
%config IPCompleter.greedy=True;
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
svr = pd.read_csv("svr_results.csv", index_col=0).rename(columns={"model": "Regresor", "name": "Dataset"})

regTree = pd.read_csv("regression_tree_results.csv", index_col=0).rename(columns={"model": "Regresor", "name": "Dataset"})

nn = pd.concat([pd.read_csv("nn_results1.csv", index_col=0),
                pd.read_csv("nn_results2.csv", index_col=0),
                pd.read_csv("nn_results3.csv", index_col=0)]).reset_index().drop(columns=['index'],axis=1).rename(columns={"model": "Regresor", "name": "Dataset"})

# Boosting methods
gradientBoostingReg = pd.read_csv("gbrResult.csv", index_col=0).rename(columns={"model": "Regresor", "name": "Dataset"})
histGradientBoostingReg = pd.read_csv("hgbrResult.csv", index_col=0).rename(columns={"model": "Regresor", "name": "Dataset"})
adaBoostingReg = pd.read_csv("abrResults.csv", index_col=0).rename(columns={"model": "Regresor", "name": "Dataset"})
randForestReg = pd.read_csv("rfrResults.csv", index_col=0).rename(columns={"model": "Regresor", "name": "Dataset"})
```

```python
#df = pd.concat([dfTree,dfNN,dfSVC]).reset_index()
ultimate = pd.concat([svr,regTree, nn, gradientBoostingReg, histGradientBoostingReg,
                      adaBoostingReg,randForestReg]).reset_index().drop(columns=['index'],axis=1)
#ultimate=ultimate.rename(columns={"model": "Classifier", "name": "Dataset"})
ultimate.head()
```

```python
def plot_heatmap(df, col, title, palette="YlGnBu",cMin=-2.0,cMax=1.0,numLabels=True):
    plt.title(title)
    ax = sns.heatmap(df.pivot(index='Dataset', columns='Regresor', values=col), annot=numLabels, cmap=palette, fmt='.3g', vmin=cMin, vmax=cMax)
    bottom, top = ax.get_ylim()
    ax.set_ylim(bottom + 0.5, top - 0.5)
    plt.show()
```

```python
plot_heatmap(ultimate, "avg_fit_time","Čas učení (s)",cMin=0,cMax=150,palette="YlOrRd", numLabels=False)
```
